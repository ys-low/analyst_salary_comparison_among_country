```python
from bs4 import BeautifulSoup
import time
import pandas as pd
import requests
```

```python
url1= "https://livingcost.org/cost"
page = requests.get(url1)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
inside_table=list(soup.tbody.find_all('tr'))
```

```python
inside_table[0].find_all('span')[0].text
inside_table[0].find_all('span')[1].text
```

```python
table_country=[]
table_cost=[]
for i in range(len(inside_table)):
    table_country.append(inside_table[i].find_all('span')[0].text),
    table_cost.append(inside_table[i].find_all('span')[1].text)
```

```python
cost_living=pd.concat((pd.DataFrame(table_country),pd.DataFrame(table_cost)),axis=1)
```

```python
cost_living.columns = ["country", "cost_living"]
```

```python
cost_living['cost_living'] = cost_living['cost_living'].str[1:]
```

```python
# cost_living.cost_living.astype(int)
cost_living['cost_living']=pd.to_numeric(cost_living['cost_living'])
```
